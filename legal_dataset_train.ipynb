```python
import os
base_path = '/content/drive/MyDrive/legal-doc-dataset'
os.chdir(base_path)

print("Current directory:", os.getcwd())
```

```python
judgement_path = os.path.join(base_path, 'IN-Ext/judgement')
summary_a1_path = os.path.join(base_path, 'IN-Ext/summary/full/A1')
summary_a2_path = os.path.join(base_path, 'IN-Ext/summary/full/A2')
```

```python
judgements_all=[]
for judgement_file in os.listdir(judgement_path):
    # Get the judgement name without the file extension
    judgement_name = os.path.splitext(judgement_file)[0]

    # Define the full paths to the judgement file and its summaries
    judgement_full_path = os.path.join(judgement_path, judgement_file)
    summary_a1_full_path = os.path.join(summary_a1_path, judgement_file)
    summary_a2_full_path = os.path.join(summary_a2_path, judgement_file)

    # Read the judgement content
    with open(judgement_full_path, 'r') as file:
        judgement_content = file.read()

    # Read the summaries from A1 and A2
    with open(summary_a1_full_path, 'r') as file:
        summary_a1_content = file.read()

    with open(summary_a2_full_path, 'r') as file:
        summary_a2_content = file.read()

    file_content = {'judgement': judgement_content,
                    'summary_a1': summary_a1_content,
                    'summary_a2': summary_a2_content}
    judgements_all.append(file_content)
    # Print the content of judgement and its summaries
    # print("Judgement:", judgement_name)
    # print("Judgement content:\n", judgement_content)
    # print("A1 Summary:\n", summary_a1_content)
    # print("A2 Summary:\n", summary_a2_content)
    # print("\n")
```

```python
import os
import json

# # Define the base path and the list of articles
# base_path = "path_to_your_directory"
# judgements_all = [...]  # Your list of articles

# Ensure the directory exists
os.makedirs(base_path, exist_ok=True)

# Function to save data to JSONL
def save_as_jsonl(data, filename):
    with open(filename, 'w') as f:
        for item in data:
            json.dump(item, f)
            f.write('\n')  # New line for JSONL format

# Save all articles to a single JSONL file
save_as_jsonl(judgements_all, os.path.join(base_path, 'all_articles.jsonl'))

print("All articles have been saved to 'all_articles.jsonl'")
```

```python
# Data loading part
data_files = {'data': os.path.join(base_path, 'all_articles.jsonl')}
dataset = load_dataset("json", data_files=data_files)

print("Dataset loaded successfully")
```
